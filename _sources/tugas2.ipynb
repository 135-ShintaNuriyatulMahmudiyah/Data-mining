```python
import pandas as pd
```

```python
dataset=("https://raw.githubusercontent.com/ApoorvRusia/Naive-Bayes-classification-on-Iris-dataset/master/iris.csv")
```

```python
df= pd.read_csv(dataset)
```

```python
df.head(10)
```

```python
# CONSTAN SERIES

SEPAL_LENGTH_SERIES = df["sepal_length"]
SEPAL_WIDTH_SERIES = df["sepal_width"]
PETAL_LENGTH_SERIES = df["petal_length"]
PETAL_WIDTH_SERIES = df["petal_width"]
```



*   Discretization adalah proses merubah atribut numerik menjadi atribut kategorikal dengan cara melakukan mapping data tersebut ke dalam interval atau concept label.
*   Teknik yang digunakan adalah :
    1.   Binning : proses topdown ini melakukan splitting berdasarkan jumlah bin
    2.   Histogram analysis: membuat partisi histogram ke dalam bucket atau bins.
    3. Cluster analysis : algoritma clustering diaplikasikan ke dalam atribut numerik dan membagi atribut ke dalam kelompok atau cluster.
    4. Decision tree anlysisi : teknik ini menentukan sebuah atribut yang memiliki entropi minimum dan kemudian dijadikan splitting point, kemudian secara rekursif melakukan partisi sehingga menghasilkan discretization yang memiliki hirarki
    5. Correlation analysis : mencari interval bersebelahan yang terbaik kemudian menggabungkan mereka ke dalam interval yang lebih besar.

*   Pendekatan Discretization
    1.   Equal-width intervals adalah discretization yang membagi data numerik menjadi beberapa kelompok dengan lebar kelompok yang kurang lebih sama besar.
    2.   Equal-frequency intervals adalah discretization yang membagi data numerik menjadi beberapa kelompok dengan jumlah anggota yang kurang lebih sama besar.
*   **Cut** adalah sebuah method pada library pandas untuk melakukan perhitungan equal width intervals











```python
# Syntax
pd.cut(series, interval, right=True,label)
```

**Sepal Width**
*   Equal Width Intervals lebar kelopak bunga Iris
*   Dikelompokkan menjadi 3 kategori yaitu kategori sedikit lebar, lebar dan sangat lebar.



```python
# equal-width intervals

labels = ["Sedikit Lebar", "Lebar", "Sangat Lebar"]
amount_of_binning = len(labels)

sepal_width_ew_binning = pd.cut(SEPAL_WIDTH_SERIES, amount_of_binning, True, labels)
labelled_sepal_width_ew_binning = sepal_width_ew_binning.value_counts()
interval_sepal_width_ew_binning = pd.cut(SEPAL_WIDTH_SERIES, amount_of_binning, True).value_counts()
```

```python
# dataframe of sepal-width and sepal category
pd.concat((SEPAL_WIDTH_SERIES, sepal_width_ew_binning), axis=1)
```

```python
# equal-width intervals binning with label
labelled_sepal_width_ew_binning
```

```python
# equal-width intervals without label
interval_sepal_width_ew_binning
```

**Qcut**
Qcut adalah sebuah method pada library pandas untuk melakukan perhitungan equal frequency intervals




**Sepal Width**
*   Equal Frequency  Intervals lebar kelopak bunga Iris
